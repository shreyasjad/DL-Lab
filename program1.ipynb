```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
plt.figure(figsize=(12,8))
def plot_sigmoid():
    x=np.linspace(-10,10,100)
    y=1/1+np.exp(-x)
    plt.subplot(2,3,1)
    plt.plot(x,y)
    plt.xlabel("value of x")
    plt.ylabel("value")
    plt.title("Sigmoid Activation function")
    plt.show()

def plot_tanh():
    x=np.linspace(-10,10,100)
    y=np.tanh(x)
    plt.subplot(2,3,2)
    plt.plot(x,y)
    plt.title("Hyperbolic Tangent (tanh) Activation Function")
    plt.xlabel("x")
    plt.ylabel("tanh(x)")
    plt.grid(True)
    plt.show()    
    
def plot_reLu():
    x=np.linspace(-10,10,100)
    y=np.maximum(0,x)
    plt.subplot(2,3,3)
    plt.plot(x,y)
    plt.xlabel("Value of x")
    plt.ylabel("reLu activation function")
    plt.grid(True)
    plt.show()

def plot_leakyreLu():
    x=np.linspace(-10,10,100)
    y=np.where(x>0,x,0.01*x)
    plt.subplot(2,3,4)
    plt.plot(x,y)
    plt.xlabel("Value of x")
    plt.ylabel("Leaky reLu activation function")
    plt.grid(True)
    plt.show()

def plot_softmax():
    x=np.linspace(-10,10,100)
    ex=np.exp(x-np.max(x))
    y=ex/np.sum(ex,axis=0)
    plt.subplot(2,3,5)
    plt.plot(x,y)
    plt.xlabel("Value of x")
    plt.ylabel("Softmax activation function")
    plt.grid(True)
    plt.show()

```

```python
plt.tight_layout()
plot_sigmoid()
plot_tanh()
plot_reLu()
plot_leakyreLu()
plot_softmax()
```
